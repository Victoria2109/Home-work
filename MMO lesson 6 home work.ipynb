Найти коэффициенты

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df= pd.read_csv("train.csv")
df
```

```python
df = df.drop(columns = 'index')
df
```

```python
Y = np.array(df['target'])
Y
```

```python
X = np.column_stack((df.drop(columns = 'target'), np.ones(len(df)))) #первый столбец из data, второй - единичный вектор длины len(data)
X
```

МНК

```python
MNK = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
MNK
```

```python
y_pred = np.dot(X, MNK)
```

```python
y_pred
```

```python
r2_score(Y, y_pred)
```

```python
 mean_absolute_percentage_error(Y, y_pred)
```

Градиентный спуск

```python
X = df.drop(columns = 'target') #первый столбец из data, второй - единичный вектор длины len(data)
X
```

```python
def add_ones(x):
    return x.insert(0,'x0', np.ones(x.shape[0]))
```

```python
#thetas - матрица коэффициентов b
def h(x, thetas):
    return np.dot(x, thetas)
```

```python
def gradient(x, y, thetas, n):
    return np.dot(-x.T, (y - h(x, thetas))) / n
```

```python
def fit(x, y, iter = 20000, learning_rate = 0.01):
    x, y = x.copy(), y.copy()
    # функцию add_ones() мы объявили выше
    
    add_ones(x)
    
    thetas, n = np.zeros(x.shape[1]), x.shape[0]
    
    for i in range(iter):
        grad = gradient(x, y, thetas, n)
        thetas -= learning_rate * grad
    return thetas
```

```python
thetas = fit(X, Y, iter = 20000, learning_rate = 0.01)
```

```python
thetas[0], thetas[1:]
```

```python
def predict(x, thetas):
    x = x.copy()
    add_ones(x)
    return np.dot(x, thetas)
```

```python
y_pred = predict(X, thetas)
```

```python
r2_score(Y, y_pred)
```

```python
 mean_absolute_percentage_error(Y, y_pred)
```
